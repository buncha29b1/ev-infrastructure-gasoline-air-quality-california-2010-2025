```python
import geopandas as gpd
import pandas as pd
```

```python
ca_counties = gpd.read_file(
    "https://www2.census.gov/geo/tiger/TIGER2023/COUNTY/tl_2023_us_county.zip"
)
ca_counties = ca_counties[ca_counties["STATEFP"] == "06"]
```
